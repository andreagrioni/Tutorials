```
# list all repository's files
## first we need to load the os python module
## we can import a specific python module
## using the function 'import'

import os

## we can new use the listdir(path) method
## to get all entries containded within
## path;
## listdir() returns a list.

entries = os.listdir('./train')
```

```
## we can get the length of entries list
## by the len method

print( len(entries) )
```

```
## or we want to print all the elements
## of the list

print( entries )
```

```
## The file train.sam contains biological sequences
## that were aligned to the human reference by
## the aligner tool STAR.
## we want to select only that file, and drop all 
## the others.

## there are several ways to do it:
# 1: list is an ordered sequence of elements,
# therefore you can access to a specific element
# by selecting its positional index in the list.
# In our case the train.sam file is in the 4rd position,
# but index starts from 0 in Python.
# Therefore the correct index position of file train.sam
# is 3.

print(entries[3])

# 2: if you do not know the index, you can loop
# over the list till you will find the element matching
# your file name and save it into a variable

for entrie in entries:
  if entrie == 'train.sam':
    target = entrie

print(f'target found: {target}')

# 3: a second way of looping over the list but 
# keeping the positional index of train.sam file
# is by using range method.

for index in range(0, len(entries)):
  if entries[index] == 'train.sam':
    print(f'the positional index {index} ' +
          f'contains the string {entries[index]}')
    
# the range(start, end, [step]) method generate a list of integer
# within the specified interval [start, end]

```
