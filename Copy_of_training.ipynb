```
## we can import a specific python module
## using the function 'import'
import shutil
import os
import psutil
import sys
from copy import copy
# clean drive if train folder already exists
if os.path.exists('train/'):
  # remove the folder and its content
  shutil.rmtree('./train/')

# copy public 'train' repository from github
! git clone https://github.com/andreagrioni/train.git
```

```
# list all repository's files
## os.listdir(path) method
## stores all entries containded within
## the path as a list;

entries = os.listdir('./train')
print( f'- the content of the list: {entries}')

## len() returns the length of the list

print( f'- number of elements in the list: {len(entries)}' )
```

```
## The file train.sam contains biological RNA sequences
## that were aligned to the human reference.
## we want to select only that file, and drop all 
## the others.

## there are several ways to do it:
# 1: list is an ordered sequence of elements,
# you can access to a specific element
# by selecting its corresponding positional index.
# In Python, index starts from 0.

# In our entries list, the train.sam file index is 3.
# use [index] To access an element of a list at
# the specific index.

# Select the content of the list enries at index 3:
print(f'1 element stored at index 3 of list entries: {entries[3]}')

# 2a: if you do not know the index, you can loop
# over the list till you will find the element matching
# your file name and save it into a variable

for entrie in entries:
  if entrie == 'train.sam':
    v_target = entrie

print(f'2a train.sam file name assigned to the variable target: {v_target}')

# 2b: a powerful Python tool is list comprehension, 
# which provides you a simple way to create list according
# to specific paramenters.

l_target = [ enrie for enrie in entries if entrie == 'train.sam' ]

print(f'2b train.sam file name stored into a new list l_target: {l_target}')

# 4: a second way of looping over the list but 
# keeping the positional index of train.sam file
# is by using range method.
# the range(start, end, [step]) method generate a list of integer
# within the specified interval [start, end]

for index in range(0, len(entries)):
  if entries[index] == 'train.sam':
    print(f'3 the positional index {index} ' +
          f'contains the string {entries[index]}')
```

```
# we want to read the content of train.sam file and perform
# some operations on it.
# we first need to set the relative path of the file in order
# to access it.

train_path = os.path.join( './train/train.sam')

# open and read the train.sam file
# The python 'with' statement is handy 
# in this case becasue we have two related
# operations, open and close file, which we
# want to execute as a pair, with a block of
# code in between 

with open(train_path, 'r') as ifile:
  lfile = ifile.readlines()
  # use the python list slide notation to: 
  # 1 print the first 10 lines:
  print(f'1 {lfile[ : 10 ]}' )
  # 2 print the last 5 lines
  print(f'1 {lfile[ -5 :  ]}' )
  # 3 print lines between index 40 and 45
  print(f'1 {lfile[ 40 : 46 ]}' )
  ## right edge is 46 because python list slide
  ## notation is inclusive for the left edge 
  ## and exclusive for the right edge.
```

```
# you may be familiar with lithe SAM file format.
# You can (should) read more here:

# https://samtools.github.io/hts-specs/SAMv1.pdf

# sam file first field is read_id
# sam file second field is a bitwise flag
# sam file third field is reference name
# sam file forth field is reference position

# we want to loop over the train.sam file
# and keep only the fields 1 to 4. These 
# fields will be stored into a new 
# list variable named 'targets'

# define relative path to train.sam:
train_path = os.path.join( './train/train.sam')

# open train.sam file and perform operations
with open(train_path, 'r') as ifile:
  # this time we are not gonna read and store
  # the whole file into a variable, but
  # we read the file line by line - so we
  # save memory.
  
  # we define two list variable 'targets':
  # you can initiate a variable list
  # by list() or []
  
  target_columns = list() 
  target_fields = list()
  
  # is equal to:
  #target_columns = [] 
  #target_fields = []
  
  for line in ifile.readlines(): 
    # readlines() method automatically
    # breaks the input file into lines 
    # anytime it meets the special 
    # character new line: '\n'.
    # Thus, each line is a string and 
    # we can apply Python string methods
    # on it.
    # Look up for 'String Methods'
    # at this URL:
    
    # https://docs.python.org/3/library/stdtypes.html
    
    # we know that SAM file is tab 
    # delimited, therefore we should 
    # break the line anytime we 
    # find the tab simbol '\t'.
    # break the line into columns
    # by using the 'split' method:
    
    columns = line.split('\t')
    
    # now, columns is a list containing
    # all the fields of line.
    # We want to keep only fields that 
    # are useful to us so we can
    # save memory. We use Python
    # slide notation to select the fields
    # of interest, and we apply it straight
    # after the list generated by the split 
    # method.
    
    fields = copy(line.split('\t')[ 0 : 4 ]) # or [ : 4 ]
    
    # we want to add each columns and fields list
    # into the lists targets_columns and targets_fields,
    # respectively. We can do it by using the 
    # list method 'append'
    
    target_columns.append( columns )
    target_fields.append( fields )
    
    # once the for loop will reach the
    # end of file (EOF) 'with' statment
    # will automatically close the file and
    # exits this code block. voila.

# I pointed out that storing only fields of 
# interest [ 1 to 4 ] is gonna save some memory.
# We can verify it by comparing the memory size of
# our lists target_columns and target_fields

# print memory size of columns
print(f'columns memory size is {sys.getsizeof( columns )}' )
# print memory size of fields
print(f'fields memory size is {sys.getsizeof( fields )}\n\n' )


print('but..')

# but if you check the memory size of
# target_columns and target_fields,
# this will be the same
if (sys.getsizeof( target_columns )
         == sys.getsizeof( target_fields ) ):
  print(f'target_columns and target_fields ' +
       f'have the same size')
  
# the two final list have the same size because
# the method getsizeof returns only the size of the
# object list, but it is not iterating within 
# the sizes of elements within the object.
# If you want to know more about it:

# search for 'sys.getsizeof'
# https://docs.python.org/3/library/sys.html?highlight=sys#module-sys
```
